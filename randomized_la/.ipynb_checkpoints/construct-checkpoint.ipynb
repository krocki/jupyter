```python
# https://amath.colorado.edu/faculty/martinss/2014_CBMS/Lectures/lecture05.pdf
# Randomized methods for low-rank approximation

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
image = np.asarray(Image.open('lena_gray.gif').convert("L"))
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
M, N = image.shape
A = image
K = 32 # ~ 1/16
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
```

```python
# standard SVD
[U, S, V] = np.linalg.svd(A)
# choose top k singular values
A1 = np.dot(U[:,:K] * S[:K], V[:K,:])
plt.imshow(A1, cmap='gray', vmin=0, vmax=255)
ref_svd = np.copy(A1)
```

```python
# Method 1
# Given an m  n matrix A
# compute an approximate rank-k SVD A  UDV

# Draw an n  k Gaussian random matrix G.
G = np.random.randn(N, K) # N x K
# Form the m  k sample matrix Y = A G.
y = np.dot(A, G) # M x K
# Form an m  k orthonormal matrix Q, st Y = QR
[Q, R] = np.linalg.qr(y) # q: M x K, r: K x K
# Form the k  n matrix B = Q A
B = np.dot(Q.T, A) # K x N
# Compute the SVD of the small matrix B: B = U D V
[U0, S, V] = np.linalg.svd(B, full_matrices = 0) # KxK, K, NxN
# Form the matrix U = QU0
U = np.dot(Q, U0)
print(U.shape, U0.shape, S.shape, V.shape)
# check
A0 = U[:, :K] @ np.diag(S[:K]) @ V[:K, :]
plt.imshow(A0, cmap='gray', vmin=0, vmax=255)
```

```python
# Method 2
# r - rank
# q - number of power iterations
# p - oversampling factor
def rSVD(X, r, q=0, p=0):
    ny = X.shape[1]
    # random projection matrix P
    P = np.random.randn(ny, r+p)
    # Z - tall skinny, MxK = MxN * NxK
    # sample column space of X with P matrix
    Z = X @ P
    # amplify singular values' decay
    for k in range(q):
        Z = X @ (X.T @ Z)
    Q, R = np.linalg.qr(Z, mode='reduced')
    Y = Q.T @ X
    UY, S, VT = np.linalg.svd(Y, full_matrices=0)
    U = Q @ UY
    return U, S, VT

[u, s, v] = rSVD(A, K, 1, 0)
A0 = u @ np.diag(s) @ v
plt.imshow(A0, cmap='gray', vmin=0, vmax=255)
```

```python
# method 3
# https://arxiv.org/pdf/1810.06860.pdf
# https://arxiv.org/pdf/0909.4061.pdf
# compression via randomized sketching?
# https://www.math.purdue.edu/~xiaj/FastSolvers2021/slides/li.pdf
def power_iteration(A, Omega, power_iter = 1):
    Y = A @ Omega
    for q in range(power_iter):
        Y = A @ (A.T @ Y)
    Q, _ = np.linalg.qr(A @ Y)
    return Q
def rsvd(A, Omega):
    Q = power_iteration(A, Omega)
    B = Q.T @ A
    u_tilde, s, v = np.linalg.svd(B, full_matrices = 0)
    u = Q @ u_tilde
    return u, s, v

omega = np.random.randn(A.shape[1], K)
[u, s, v] = rsvd(A, omega)
A0 = u[:, :K] @ np.diag(s[:K]) @ v[:K, :]
plt.imshow(A0, cmap='gray', vmin=0, vmax=255)
```

```python
# FFT
r,c = A.shape
keep = K / r

im_fft = np.fft.fft2(A)
im_fft_copy = im_fft.copy()

im_fft_copy[int(r*keep):int(r*(1-keep)),:] = 0
im_fft_copy[:,int(c*keep):int(c*(1-keep))] = 0

D = np.fft.ifft2(im_fft_copy).real

plt.imshow(D, cmap='gray', vmin=0, vmax=255)
```

```python
# QR
[Q, R] = np.linalg.qr(A)
# set some R rows to 0
R[45:64, :] = 0
A3 = np.dot(Q, R)
print(Q.shape, R.shape)
plt.imshow(A3, cmap='gray', vmin=0, vmax=255)
```

```python
m, n = A.shape
c = 100 # number of samples

colnorm = np.linalg.norm(A, axis=0)
sumnorm = np.sum(colnorm)
pcol = colnorm / sumnorm

samples = np.random.choice(n, c, replace=True, p=pcol)

C = np.zeros((m, c))

for t in range(c):
    it = samples[t]
    C[:, t] = A[:, it] / np.sqrt(c * pcol[it])
    
[U, S, V] = np.linalg.svd(np.dot(C.T, C))
sigma = np.sqrt(np.diag(S[:K]))

H = np.zeros((m, K))
for t in range(K):
    h = np.dot(C, U[:, t])
    nh = np.linalg.norm(h)
    if nh == 0:
        H[:, t] = np.zeros((H, 1))
    else:
        H[:, t] = h/nh;

invsigma = np.diag(1/sigma[sigma>0])
V = np.dot(invsigma, np.dot(H.T, A))
A0 = np.dot(np.dot(H, sigma), V)
plt.imshow(A0, cmap='gray', vmin=0, vmax=255)
```

```python
import scipy.linalg as la
import scipy.linalg.interpolative as sli

# Rank-Revealing Truncated QR
epsilon = 1e-3
def qr_decomp(A):
    q, r, p = la.qr(A, pivoting=True, mode="economic")
    diag_r = np.diag(r)
    r = r[:, np.argsort(p)]
    
    flags = np.abs(diag_r) >= epsilon
    q = q[:, flags]
    r = r[flags]
    return q, r, np.abs(diag_r)

def id_decomp(A):
    k, idx, proj = sli.interp_decomp(A, epsilon)
    sort_idx = np.argsort(idx)
    B = A[:,idx[:k]]
    P = np.hstack([np.eye(k), proj])[:,np.argsort(idx)]
    return B, P

[Q, R, P] = qr_decomp(A)
A0 = Q @ R
plt.imshow(A0, cmap='gray', vmin=0, vmax=255)
```

```python
[B, P] = id_decomp(A.astype(np.float64))
A0 = B @ P
plt.imshow(A0, cmap='gray', vmin=0, vmax=255)
```
